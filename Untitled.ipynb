```python
import pandas as pd
import numpy as np

def make_sources(N=6000):
    N = 6000
    x = np.arange(N)
    y = 5000 + x*1.3
    x0 = x - 0.5
    x1 = x + 0.5
    yhi = y + (100 + 30*np.random.random(size=N))
    ylo = y - (100 + 30*np.random.random(size=N))
    
    my_val = y + np.random.normal(size=N)*120
    my_hi = my_val + 150
    my_lo = my_val - 100
    my_center = (my_hi + my_lo)/2
    my_height = my_hi - my_lo

    good = (my_lo < yhi) & (my_hi > ylo)
    bad = ~good

    source = ColumnDataSource(data=dict(x=x, y=y, x0=x0, x1=x1,
                                       my_val=my_val, yhi=yhi, ylo=ylo,
                                        my_hi=my_hi, my_lo=my_lo,
                                       my_height=my_height, my_center=my_center))
    
    out_source = ColumnDataSource(data=dict(x=x[bad], y=y[bad], x0=x0[bad], x1=x1[bad],
                                       my_val=my_val[bad], yhi=yhi[bad], ylo=ylo[bad],
                                        my_hi=my_hi[bad], my_lo=my_lo[bad],
                                       my_height=my_height[bad], my_center=my_center[bad]))
    
    return source, out_source

def make_plot(source, out_source, prop='y'):
    p = figure(plot_width=1000, plot_height=400)
    
    p.lod_factor = 10
    p.xaxis.major_label_text_color = None
    p.xaxis.major_tick_line_color = None
    p.xaxis.minor_tick_line_color = None
    p.xgrid.grid_line_color = None
    
    
    err_x = np.concatenate([source.data['x'], source.data['x'][::-1]])
    err_y = np.concatenate([source.data['yhi'], source.data['ylo'][::-1]])

    p.patch(err_x, err_y, alpha=0.2, color='navy')
    p.segment(x0='x0', x1='x1', y0='y', y1='y', source=source)

    p.rect(x='x', y='my_center', height='my_height', width=0.3, source=source,
          color='black', alpha=0.2)
    p.rect(x='x', y='my_center', height='my_height', width=0.35, source=out_source,
          color='red', alpha=0.35)

    #p.circle(x='x', y='my_val', source=source, radius=0.4, color='black', alpha=0.5)
    p.circle(x='x', y='my_val', source=out_source, radius=0.4, color='red', alpha=0.5)
    
    return p
    
```
